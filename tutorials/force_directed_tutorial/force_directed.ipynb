# Force-Directed Graph Generation Example

In this short tutorial, we show an example of automatic diagram generation for a small circuit using FlyBrainLab using a force-directed layout. The code shown in this tutorial is intended to be a possible starting point for diagram design.

**Important:** Before starting to run the code below, **(i)** create an FBL workspace for Hemibrain, **(ii)** change the kernel for this notebook to the kernel you just created.

Let us start by querying the neurons for which we will be constructing the diagram:

```python
from gen_diagram import * # Import our graph generation utility

client = fbl.get_client()
# Let's pick two subcircuits connected to one another
_res = client.executeNLPquery('show $VP3$') 
_res = client.executeNLPquery('add $VP5$')
```

Then let us create the diagram visualization:

```python
generate_graph(client) # Generates a graph for the current client
```

This graph has various features: (i) hovering on neurons highlights them; (ii) double clicking on them reveals a number of interesting features about the neurons, including the number of neurons they are connecting to, (iii) clicking on them changes their visibility. Interested users can create their own graph visualization functions by altering the loadrawGEXF and initializeGEXF functions and loading new versions of it.
