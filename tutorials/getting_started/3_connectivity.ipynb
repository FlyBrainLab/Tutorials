# Introduction to FBL: Part 3: Connectivity

In this example, we show how you can build a dendrogram of your current circuit. For this example, start by pressing the 'Create FBL Workspace' button and selecting FlyCircuit.

First, let us start by querying a sufficiently small circuit:

```python
my_client = fbl.get_client()
res = my_client.executeNLPquery('show GABAergic neurons in the ellipsoid body')
```

We can then get a connectivity dendrogram simply by running:

```python
%matplotlib inline
my_client.getConnectivityDendrogram()
```

This generates a connectivity matrix by querying FFBO and 
builds an internal representation of the circuit.

Secondly, it generates a connectivity matrix:

```python
M = my_client.getConnectivityMatrix()
```

Finally, it uses the seaborn library to generate a nice visualization:

```python
import pandas as pd
import seaborn as sns
M = pd.DataFrame(M, index = my_client.out_nodes, columns = my_client.out_nodes)
sns.clustermap(M)
```

You can look at the nodes of the generated circuit by looking at the output of the cell below:

```python
# Refer to 
# https://github.com/FlyBrainLab/FBLClient/blob/433be22b06b89cd948569c2de1a62057af1ccd7d/flybrainlab/Client.py#L2000
# for details of the circuit construction.
my_client.C.G.nodes() # my_client.C is an execution-ready Neuroballad circuit
```
